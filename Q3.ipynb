## Rewards Reciept Item List Data

```python
# Copy the receipts data into new dataframe
r = receipts_df.copy()
```

```python
# Remove Null rows
r_clean = r.dropna(subset=['rewardsReceiptItemList'])
```

```python
# Create a new dataframe for all the receipt items
rewardsReceiptItems = pd.DataFrame()
for i in r_clean.index:
  row = pd.DataFrame(r_clean['rewardsReceiptItemList'][i])
  rewardsReceiptItems = rewardsReceiptItems.append(row, ignore_index = True)
```

```python
rewardsReceiptItems.head()
```

## Data Quality Evaluation
1. 4011 barcode has a description of ITEM NOT FOUND	which means that there was no brand identified while scanning
2. brandCode can be used as a foreign key to the Brands table




```python
# Information of Rewards Reciept items
rewardsReceiptItems.info()
```

```python
# Percentage of rows in rewardsreceiptitems data per column which are null
(1 - rewardsReceiptItems.count() / len(rewardsReceiptItems)) * 100
```



> Data Quality Evaluation


1.  brandCode has ~62% null values which makes it difficult to match the brand name with the item scanned from reciepts
2. pointsEarned has ~86% null values which makes it difficult to maintain the points earned through receipt for a consumer
2.   barcode has ~55% null values which makes it difficult to identify the item and link it to the appropriate brand
3. 2% of item price values are null. This makes it difficult to populate points earned for those consumers
